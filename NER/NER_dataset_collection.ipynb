```python
from datasets import load_dataset, DatasetDict, concatenate_datasets, Dataset
import random
import matplotlib.pyplot as plt

# List of languages
languages = {
    "yo": "Yoruba",
    "ar": "Arabic",
    "zh": "Chinese",
    "ru": "Russian",
    "hi": "Hindi",
    "ja": "Japanese",
    "sw": "Swahili",
    "bn": "Bengali",
    "tr": "Turkish"
}

# Load datasets
datasets = {}
min_len = float('inf')

for code in languages:
    dataset = load_dataset("wikiann", code)
    datasets[code] = dataset
    lang_min_len = min(len(dataset["train"]), len(dataset["validation"]), len(dataset["test"]))
    min_len = min(min_len, lang_min_len)  # Track minimum length for balance

print(f"Balanced sample size per split per language: {min_len}")

# Sample and combine balanced datasets
train_datasets = []
test_datasets = []

language_counts = {}

for code in languages:
    # Shuffle before sampling
    train_split = datasets[code]["train"].shuffle(seed=42).select(range(min_len))
    test_split = datasets[code]["test"].shuffle(seed=42).select(range(min_len))

    # Add language tag for tracking
    train_split = train_split.add_column("language", [languages[code]] * min_len)
    test_split = test_split.add_column("language", [languages[code]] * min_len)

    train_datasets.append(train_split)
    test_datasets.append(test_split)

    language_counts[languages[code]] = min_len

# Concatenate into unified datasets
final_train = concatenate_datasets(train_datasets)
final_test = concatenate_datasets(test_datasets)

# Save to disk
final_dataset = DatasetDict({
    "train": final_train,
    "test": final_test
})
final_dataset.save_to_disk("/content/drive/MyDrive/Tokenizer_New/NER_dataset/balanced_multilingual_ner")

print(" Balanced multilingual dataset saved to 'balanced_multilingual_ner'")

# Plot language distribution
plt.figure(figsize=(10, 5))
plt.bar(language_counts.keys(), language_counts.values(), color="skyblue")
plt.title("Language Distribution in Balanced Multilingual Dataset")
plt.xlabel("Language")
plt.ylabel("Number of Samples (per split)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from datasets import Dataset, DatasetDict

def parse_conll_txt(filename):
    sentences = []
    tokens = []
    ner_tags = []

    with open(filename, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                if tokens:
                    sentences.append({"tokens": tokens, "ner_tags": ner_tags})
                    tokens = []
                    ner_tags = []
            else:
                if ' ' in line:
                    token, tag = line.rsplit(' ', 1)
                    tokens.append(token)
                    ner_tags.append(tag)
                else:
                    # In case a line is broken (only one word), treat it as non-entity
                    tokens.append(line)
                    ner_tags.append("O")

        # Catch the last sentence if the file doesn't end with a newline
        if tokens:
            sentences.append({"tokens": tokens, "ner_tags": ner_tags})

    return sentences
```

```python
from datasets import load_dataset, DatasetDict, concatenate_datasets, Dataset, ClassLabel, Sequence, Value, Features
import matplotlib.pyplot as plt

# Step 1: Define languages
languages = {
    "yo": "Yoruba",
    "ar": "Arabic",
    "zh": "Chinese",
    "ru": "Russian",
    "hi": "Hindi",
    "ja": "Japanese",
    "sw": "Swahili",
    "bn": "Bengali",
    "tr": "Turkish"
}

# Step 2: Load WikiANN datasets (except Yoruba)
datasets = {}
min_len = float("inf")

for code in languages:
    if code != "yo":
        dataset = load_dataset("wikiann", code)
        datasets[code] = dataset
        lang_min_len = min(len(dataset["train"]), len(dataset["validation"]), len(dataset["test"]))
        min_len = min(min_len, lang_min_len)

print(f" Balanced sample size per split per language (except Yoruba): {min_len}")

# Step 3: Get label list from WikiANN to map ner_tags
label_list = datasets["ar"]["train"].features["ner_tags"].feature.names
label2id = {label: i for i, label in enumerate(label_list)}
allowed_labels = set(label2id.keys())

# Step 4: Convert MasakhaneNER ner_tags to integers (FILTER unsupported tags)
def encode_labels(example):
    if all(tag in allowed_labels for tag in example["ner_tags"]):
        example["ner_tags"] = [label2id[tag] for tag in example["ner_tags"]]
        return example
    else:
        return None  # Drop unsupported tag examples

masakhaner_yoruba["train"] = masakhaner_yoruba["train"].map(encode_labels).filter(lambda x: x is not None)
masakhaner_yoruba["test"] = masakhaner_yoruba["test"].map(encode_labels).filter(lambda x: x is not None)

# Step 5: Ensure matching features
features = Features({
    "tokens": Sequence(Value("string")),
    "ner_tags": Sequence(ClassLabel(names=label_list)),
    "language": Value("string")
})

masakhaner_yoruba["train"] = masakhaner_yoruba["train"].cast(features)
masakhaner_yoruba["test"] = masakhaner_yoruba["test"].cast(features)

# Step 6: Combine and balance all datasets
train_datasets = []
test_datasets = []
language_counts = {}

for code in languages:
    lang_name = languages[code]

    if code == "yo":
        combined = concatenate_datasets([masakhaner_yoruba["train"], masakhaner_yoruba["test"]])
    else:
        combined = concatenate_datasets([datasets[code]["train"], datasets[code]["test"]])

    # Shuffle combined and take min_len * 2 samples (ensures balance)
    combined = combined.shuffle(seed=42).select(range(min_len * 2))

    # Split into 80% train, 20% test
    split = combined.train_test_split(test_size=0.2, seed=42)
    train_split = split["train"]
    test_split = split["test"]

    # Remove language column if exists
    if "language" in train_split.column_names:
        train_split = train_split.remove_columns("language")
    if "language" in test_split.column_names:
        test_split = test_split.remove_columns("language")

    # Add language column
    train_split = train_split.add_column("language", [lang_name] * len(train_split))
    test_split = test_split.add_column("language", [lang_name] * len(test_split))

    train_datasets.append(train_split)
    test_datasets.append(test_split)

    language_counts[lang_name] = len(train_split)

# Step 7: Concatenate
final_train = concatenate_datasets(train_datasets)
final_test = concatenate_datasets(test_datasets)

final_dataset = DatasetDict({
    "train": final_train,
    "test": final_test
})

# Step 8: Save to disk
save_path = "/content/drive/MyDrive/Tokenizer_New/NER_dataset/balanced_multilingual_ner"
final_dataset.save_to_disk(save_path)

print(f" Saved final dataset to: {save_path}")

# Step 9: Plot distribution
plt.figure(figsize=(10, 5))
plt.bar(language_counts.keys(), language_counts.values(), color="skyblue")
plt.title("Language Distribution in Balanced Multilingual Dataset")
plt.xlabel("Language")
plt.ylabel("Number of Samples (per split)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#  Step 9: Print dataset statistics
from collections import defaultdict

def compute_stats(split):
    lang_counter = defaultdict(int)
    for example in split:
        lang_counter[example["language"]] += 1
    return dict(lang_counter)

train_stats = compute_stats(final_dataset["train"])
test_stats = compute_stats(final_dataset["test"])

print("\n Final Dataset Statistics:")
print(" Train Samples:")
for lang in sorted(train_stats):
    print(f"   {lang:10s} : {train_stats[lang]:>5}")

print(" Test Samples:")
for lang in sorted(test_stats):
    print(f"   {lang:10s} : {test_stats[lang]:>5}")
```

```python
import os
import csv
import torch
import warnings
import evaluate
from datasets import load_from_disk
from sklearn.exceptions import UndefinedMetricWarning
from torch.utils.data import Dataset
from transformers import (
    AutoTokenizer,
    BertForTokenClassification,
    Trainer,
    TrainingArguments,
    TrainerCallback,
    logging as hf_logging
)
from collections import Counter
import numpy as np

# === Setup ===
warnings.filterwarnings("ignore", category=UndefinedMetricWarning)
hf_logging.set_verbosity_error()

base_path = "/content/drive/MyDrive/Tokenizer_New"
dataset_path = os.path.join(base_path, "NER_dataset", "balanced_multilingual_ner")
accuracy_log_file = os.path.join(base_path, "NER_models", "accuracy_log.csv")

tokenizer_sizes = ["small", "medium", "large"]
tokenizer_types = ["hf_bpe_hf", "hf_wordpiece_hf", "sp_unigram_hf"]

# tokenizer_sizes = ["small"]
# tokenizer_types = [ "hf_wordpiece_hf"]

# === Load Dataset ===
dataset = load_from_disk(dataset_path)
train_data = dataset["train"]
test_data = dataset["test"]

# === Build Label Maps from Dataset Features (No manual rebuild) ===
label_list = train_data.features["ner_tags"].feature.names
label2id = {label: i for i, label in enumerate(label_list)}
id2label = {i: label for i, label in enumerate(label_list)}
num_labels = len(label_list)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# === Compute Class Weights ===
all_labels = []
for example in train_data:
    all_labels.extend(example["ner_tags"])
label_counts = Counter(all_labels)
total_count = sum(label_counts.values())
class_weights = [total_count / (num_labels * label_counts[i]) for i in range(num_labels)]
class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)
print(" Class Weights:", {id2label[i]: round(w.item(), 3) for i, w in enumerate(class_weights)})

# === Dataset Class ===
class NerDataset(Dataset):
    def __init__(self, data, tokenizer, label2id, max_length=128):
        self.data = data
        self.tokenizer = tokenizer
        self.label2id = label2id
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]
        tokens, tags = item["tokens"], item["ner_tags"]

        encoding = self.tokenizer(
            tokens,
            is_split_into_words=True,
            max_length=self.max_length,
            return_offsets_mapping=True,
            truncation=True,
            padding="max_length"
        )

        labels = [-100] * len(encoding["input_ids"])
        word_ids = encoding.word_ids()

        # Assign label to all subwords
        for i, word_id in enumerate(word_ids):
            if word_id is None:
                continue
            labels[i] = tags[word_id]  # Already numeric IDs from dataset

        encoding["labels"] = labels
        encoding.pop("offset_mapping")
        return {k: torch.tensor(v) for k, v in encoding.items()}

# === Load seqeval metric for NER ===
metric = evaluate.load("seqeval")

def compute_metrics(p):
    preds = p.predictions.argmax(-1)
    labels = p.label_ids

    true_predictions, true_labels = [], []
    for pred_seq, label_seq in zip(preds, labels):
        pred_labels = []
        gold_labels = []
        for pred, lab in zip(pred_seq, label_seq):
            if lab != -100:
                pred_labels.append(id2label[pred])
                gold_labels.append(id2label[lab])
        true_predictions.append(pred_labels)
        true_labels.append(gold_labels)

    results = metric.compute(predictions=true_predictions, references=true_labels)
    return {
        "precision": results["overall_precision"],
        "recall": results["overall_recall"],
        "f1": results["overall_f1"],
        "accuracy": results["overall_accuracy"],
    }

# === Callback to print metrics after each evaluation ===
class PrintMetricsCallback(TrainerCallback):
    def on_evaluate(self, args, state, control, metrics=None, **kwargs):
        if metrics:
            print("\n Evaluation Results:")
            for key, value in metrics.items():
                if isinstance(value, float):
                    print(f"  {key}: {value:.4f}")

# === Custom Trainer with Class Weights ===
from transformers import Trainer
import torch.nn as nn

class WeightedLossTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False, num_items_in_batch=None):
        labels = inputs.get("labels")
        outputs = model(**inputs)
        logits = outputs.get("logits")
        loss_fct = nn.CrossEntropyLoss(weight=class_weights, ignore_index=-100)
        loss = loss_fct(logits.view(-1, self.model.config.num_labels), labels.view(-1))
        return (loss, outputs) if return_outputs else loss


# === CSV Logger Setup ===
os.makedirs(os.path.dirname(accuracy_log_file), exist_ok=True)
if not os.path.isfile(accuracy_log_file):
    with open(accuracy_log_file, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow([
            "Model Name", "Epoch",
            "Train Precision", "Train Recall", "Train F1", "Train Accuracy",
            "Test Precision", "Test Recall", "Test F1", "Test Accuracy"
        ])

# === Main Training Loop ===
for size in tokenizer_sizes:
    for tok_type in tokenizer_types:
        model_name = f"{size}_{tok_type}_NER"
        tokenizer_path = os.path.join(base_path, "vocab_final", f"vocab_final{size}", tok_type)
        model_output_dir = os.path.join(base_path, "NER_models", model_name)

        print(f"\n Running: {model_name}")
        tokenizer = AutoTokenizer.from_pretrained(tokenizer_path, local_files_only=True)
        train_dataset = NerDataset(train_data, tokenizer, label2id)
        test_dataset = NerDataset(test_data, tokenizer, label2id)

        model = BertForTokenClassification.from_pretrained(
            "bert-base-cased",
            num_labels=num_labels,
            label2id=label2id,
            id2label=id2label,
            ignore_mismatched_sizes=True
        )
        model.resize_token_embeddings(len(tokenizer))
        model.to(device)

        args = TrainingArguments(
            output_dir=model_output_dir,
            per_device_train_batch_size=16,
            per_device_eval_batch_size=16,
            num_train_epochs=3,
            eval_strategy="epoch",
            logging_strategy="epoch",
            learning_rate=5e-5,
            logging_dir=os.path.join(model_output_dir, "logs"),
            save_strategy="no",
            report_to=[],
            disable_tqdm=False
        )

        trainer = WeightedLossTrainer(
            model=model,
            args=args,
            train_dataset=train_dataset,
            eval_dataset=test_dataset,
            tokenizer=tokenizer,
            compute_metrics=compute_metrics,
            callbacks=[PrintMetricsCallback()]
        )

        trainer.train()
        train_metrics = trainer.evaluate(train_dataset)
        test_metrics = trainer.evaluate(test_dataset)

        with open(accuracy_log_file, 'a', newline='', encoding='utf-8') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow([
                model_name, 3,
                round(train_metrics.get('eval_precision', 0.0), 4),
                round(train_metrics.get('eval_recall', 0.0), 4),
                round(train_metrics.get('eval_f1', 0.0), 4),
                round(train_metrics.get('eval_accuracy', 0.0), 4),
                round(test_metrics.get('eval_precision', 0.0), 4),
                round(test_metrics.get('eval_recall', 0.0), 4),
                round(test_metrics.get('eval_f1', 0.0), 4),
                round(test_metrics.get('eval_accuracy', 0.0), 4),
            ])

        model.save_pretrained(model_output_dir)
        tokenizer.save_pretrained(model_output_dir)
        torch.cuda.empty_cache()

print("\n All NER training runs complete.")
```
